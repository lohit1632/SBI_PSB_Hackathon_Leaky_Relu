```python
columns_to_convert = ['AVERAGE_ACCT_AGE1', 'CREDIT_HISTORY_LENGTH1'] # Assuming these are the columns

def convert_yrs_mon_to_months(duration_str):
    """
    Converts a string like 'Xyrs Ymon' to the total number of months.
    Returns None for invalid formats or missing values.
    """
    if pd.isna(duration_str):
        return None
    try:
        parts = duration_str.replace('yrs', '').replace('mon', '').split()
        years = int(parts[0]) if parts[0] else 0
        months = int(parts[1]) if len(parts) > 1 and parts[1] else 0
        return years * 12 + months
    except (ValueError, IndexError):
        # Handle cases that don't match the expected format
        return None

# Apply the conversion function to the specified columns
for col in columns_to_convert:
    if col in df_filtered.columns:
        df_filtered[col] = df_filtered[col].apply(convert_yrs_mon_to_months)
        # Convert the column to a numeric type after applying the function
        df_filtered[col] = pd.to_numeric(df_filtered[col])

# Verify the data types and some values
print(df_filtered[columns_to_convert].dtypes)
print(df_filtered[columns_to_convert].head())
```

```python
import numpy as np
import pandas as pd
from scipy.stats import skew

def smart_impute(df):
    df_imputed = df.copy()

    for col in df.columns:
        if df[col].isnull().sum() == 0:
            continue  # Skip if no missing

        if df[col].dtype == 'object' or df[col].dtype.name == 'category':
            # Categorical: use mode
            mode_val = df[col].mode().iloc[0]
            df_imputed[col].fillna(mode_val, inplace=True)

        elif np.issubdtype(df[col].dtype, np.number):
            # Numeric: check skewness to decide mean or median
            col_skew = skew(df[col].dropna())
            if abs(col_skew) > 1:  # highly skewed
                median_val = df[col].median()
                df_imputed[col].fillna(median_val, inplace=True)
            else:
                mean_val = df[col].mean()
                df_imputed[col].fillna(mean_val, inplace=True)

        else:
            # Other types (rare): fallback to mode
            df_imputed[col].fillna(df[col].mode().iloc[0], inplace=True)

    return df_imputed
```

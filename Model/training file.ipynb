```python
columns_to_convert = ['AVERAGE_ACCT_AGE1', 'CREDIT_HISTORY_LENGTH1'] # Assuming these are the columns

def convert_yrs_mon_to_months(duration_str):
    """
    Converts a string like 'Xyrs Ymon' to the total number of months.
    Returns None for invalid formats or missing values.
    """
    if pd.isna(duration_str):
        return None
    try:
        parts = duration_str.replace('yrs', '').replace('mon', '').split()
        years = int(parts[0]) if parts[0] else 0
        months = int(parts[1]) if len(parts) > 1 and parts[1] else 0
        return years * 12 + months
    except (ValueError, IndexError):
        # Handle cases that don't match the expected format
        return None

# Apply the conversion function to the specified columns
for col in columns_to_convert:
    if col in df_filtered.columns:
        df_filtered[col] = df_filtered[col].apply(convert_yrs_mon_to_months)
        # Convert the column to a numeric type after applying the function
        df_filtered[col] = pd.to_numeric(df_filtered[col])

# Verify the data types and some values
print(df_filtered[columns_to_convert].dtypes)
print(df_filtered[columns_to_convert].head())
```

```python
import numpy as np
import pandas as pd
from scipy.stats import skew

def smart_impute(df):
    df_imputed = df.copy()

    for col in df.columns:
        if df[col].isnull().sum() == 0:
            continue  # Skip if no missing

        if df[col].dtype == 'object' or df[col].dtype.name == 'category':
            # Categorical: use mode
            mode_val = df[col].mode().iloc[0]
            df_imputed[col].fillna(mode_val, inplace=True)

        elif np.issubdtype(df[col].dtype, np.number):
            # Numeric: check skewness to decide mean or median
            col_skew = skew(df[col].dropna())
            if abs(col_skew) > 1:  # highly skewed
                median_val = df[col].median()
                df_imputed[col].fillna(median_val, inplace=True)
            else:
                mean_val = df[col].mean()
                df_imputed[col].fillna(mean_val, inplace=True)

        else:
            # Other types (rare): fallback to mode
            df_imputed[col].fillna(df[col].mode().iloc[0], inplace=True)

    return df_imputed
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore # Import zscore function from scipy.stats

def remove_outliers_zscore(df, threshold=3):
    """
    Removes outliers from a DataFrame based on the Z-score.
    Applied only to numerical columns.

    Args:
        df (pd.DataFrame): The input DataFrame.
        threshold (float): The absolute Z-score threshold. Rows with absolute
                           Z-score greater than this threshold are removed.

    Returns:
        pd.DataFrame: The DataFrame with outliers removed.
    """
    df_cleaned = df.copy() # Create a copy to avoid modifying the original DataFrame

    # Identify numerical columns
    numerical_cols = df_cleaned.select_dtypes(include=np.number).columns

    print(f"Shape before outlier removal: {df_cleaned.shape}")

    # Create a mask to keep all rows initially
    # We will update this mask to exclude rows with outliers
    combined_mask = pd.Series(True, index=df_cleaned.index)


    for col in numerical_cols:
        # Calculate Z-scores for the column, ignoring NaNs
        # .dropna() is important before calculating zscore to avoid issues with NaNs
        # and the zscore function handles NaNs by returning NaN, which is fine for the mask
        z_scores = np.abs(zscore(df_cleaned[col].dropna()))

        # Identify indices where the absolute z-score is greater than the threshold
        outlier_indices_for_col = df_cleaned[col].dropna().index[z_scores > threshold]

        # Update the combined mask: set to False for any index identified as outlier
        combined_mask[outlier_indices_for_col] = False

        print(f"  After processing '{col}': identified {len(outlier_indices_for_col)} potential outliers based on Z-score > {threshold}")


    # Filter the DataFrame using the combined mask
    # This removes a row if it was identified as an outlier in *any* numerical column
    df_cleaned_final = df_cleaned[combined_mask]

    print(f"Shape after outlier removal (Z-score > {threshold}): {df_cleaned_final.shape}")
    print(f"Total rows removed: {df_cleaned.shape[0] - df_cleaned_final.shape[0]}")


    return df_cleaned_final

# Apply the Z-score outlier removal function to df1
# You can adjust the threshold parameter if needed, e.g., threshold=2 or threshold=3.5
df_no_outliers_zscore = remove_outliers_zscore(df1, threshold=7)

# Now df_no_outliers_zscore contains the DataFrame with outliers removed based on Z-score
# You can continue your analysis with df_no_outliers_zscore
```
